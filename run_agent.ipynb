# LangChain Agent 工具调用实验

本 notebook 演示如何使用模块化的 agent_project 包来创建和运行带工具的 agent。

```python
from agent_project import create_agent

agent = create_agent()
print("Agent 创建成功！")
```

## 测试 1：天气查询工具

agent 会识别出需要调用 `get_weather` 工具来回答天气问题。

```python
result = agent.invoke({"messages": [("user", "北京今天天气怎么样？")]})

for msg in result["messages"]:
    print(f"[{msg.type}] {msg.content}")
```

## 测试 2：计算器工具

agent 会调用 `calculate` 工具来处理数学计算。

```python
result = agent.invoke({"messages": [("user", "帮我算一下 (123 + 456) * 7 等于多少")]})

for msg in result["messages"]:
    print(f"[{msg.type}] {msg.content}")
```

## 测试 3：时间查询工具

agent 会调用 `get_current_time` 工具（无参数工具的示例）。

```python
result = agent.invoke({"messages": [("user", "现在几点了？")]})

for msg in result["messages"]:
    print(f"[{msg.type}] {msg.content}")
```

## 测试 4：多工具组合调用

一个问题可能触发多个工具调用。

```python
result = agent.invoke({"messages": [("user", "现在几点了？另外上海天气如何？")]})

for msg in result["messages"]:
    print(f"[{msg.type}] {msg.content}")
```
